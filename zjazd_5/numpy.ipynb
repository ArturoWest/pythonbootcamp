### Stwórz liste sześcianów liczb od 0 do 9

```python
a = np.arange(10)
a = a**3
```

### wybierz z listy zaznaczone elementy

[ 0, 1, **8**, **27**, **64**, 125, 216, 343, 512, 729]

    

```python
a[2:5]
```

[ **0**, 1, **8**, 27, **64**, 125, 216, 343, 512, 729]

```python
a[0:6:2]
```

### przekształć by otrzymać wynik

[-1000, 1,  -1000, 27, -1000,   125,   216,   343,   512, 729]

```python
a[0:5:2]=-1000
a
```

### Odwróć listę

```python
a[::-1]
```

### W pętli for wyprintuj dla każdego elementu pierwiastek sześcienny z elementy

```python
for i in a:
    print(i ** (1/3))
```

### Czarno biały filtr

```python
im1[:,:,1] = im1[:,:,0]
im1[:,:,2] = im1[:,:,0]
plt.imshow(im1)
```

```python
suma_z_warstw = im1.sum(axis=2) #, keepims = True)
suma_z_warstw = suma_z_warstw / np.max(suma_z_warstw)
data = np.ones(3) * suma_z_warstw
data.show
plt.imshow(data)
```
